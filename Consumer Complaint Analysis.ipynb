We can perform this function in following two ways, with function. For serive issue we will do it like in this way.  

```python
def getcomp(word):
    x=df[df["Customer Complaint"].str.contains(word)].count()["Ticket #"]
    return x
```

```python
service_issue_1=getcomp("service")
```

For service_issue_2 we will do it in following way, if we don't want to use <i> str.contains() </i> function of pandas.

```python
def n_getcomp(word):
    count=0
    for x in df["Customer Complaint"]:
        if word in x:
            count=count+1
    return count
```

```python
service_issue_2=n_getcomp("customer")
```

```python
total_service_issue=service_issue_1+service_issue_2
```

```python
total_service_issue
```

Next task is to provide the state wise status of complaint in a stacked bar chart.

```python
df.groupby(by=["State","newstatus"]).size()
```

To get it in better readable form we will use <i> unstack </i> function.
